```python
#REQUIRED LIBRARIES
import csv #file creation on system
import smtplib #for sending mail
import getpass #for hiding input
import random #creating passkey
from email.mime.text import MIMEText #including subject while sending mail.
from IPython.display import display#DISPLAY's image
from IPython.display import Image#diplays' only image
import mysql.connector as sql #mysql block
from datetime import datetime #for sending date & time as input in server 
from os import system#---
from os import mkdir#----for GREETING BLOCK
from os import remove#---
from os import rmdir#----
from time import sleep #creating delay in program
from tabulate import tabulate #for arrangement
print("LIBRARIES CHECK.")
```

```python
#ALL FUNCTIONS
def greet(x): #GREETING BLOCK(SENDS AUDIO OUTPUT)
    a=x
    if len(a)==0:
        print("*DEFAULT MESSAGE IS \n BEING USED")
        a="HELLO WELCOME BACK , NICE TO MEET YOU "
        print("*MESSAGE IS>>","''",a,"''")

    b='''Dim speaks, speech
    speaks="'''
    c=b+a+'''"\n'''
    d='''set speech=CreateObject("sapi.spvoice")
    speech.speak speaks'''
    c=c+d
    mkdir("d:\piygreetdemo")
    f1=open("D:\\piygreetdemo\\WELCOME.vbs","w")
    f1.write(c)
    f1.close()
    system("D:\\piygreetdemo\\WELCOME.vbs")
    remove ("d:\piygreetdemo\welcome.vbs")
    rmdir("d:\piygreetdemo")

def seat_type(seat): #seat type finder
    a=seat
    b=int((a/3)+0.9)           #block number  
    c=0
    for i in range(1,10):
        if(b==4*i or b==(4*i)-1):
            c=2
            break;
        else:
            c=1
#c indicates 4 different algorithms for window seats
#allocating all 4 values of d to their block no. of one coach
#when c=1
    if(c==1):
         for i in range(1,18,2):
                if(b==2*i):
                    d=2
                    break;
                else:
                    d=1
    #when c=2
    if(c==2):
         for i in range(1,27):
                if(b==4*i):
                    d=4
                    break;
                else:
                    d=3
#giving seats info to >>d variable.
    if(d==1):
        if(b*3==a):
            type="AS"
        elif((b*3)-1==a):
            type="MS"
        else:
            type="WS"
    elif(d==2):
        if(b*3==a):
            type="WS"
        elif((b*3)-1==a):
            type="MS"
        else:
            type="AS"
    elif(d==3):
        if(b*3==a):
            type="AS"
        elif((b*3)-1==a):
            type="MS"
        else:
            type="WS"
    else:
        if(b*3==a):
            type="WS"
        elif((b*3)-1==a):
            type="MS"
        else:
            type="AS"
    return type;
def filename(): #creating csv file with heading
    file=open("d:\\project_train.csv","w",newline="")
    writer=csv.writer(file)
    field=["SEAT NO.","SEAT TYPE","NAME","AGE","GMAIL-ID"]
    writer.writerow(field)
    file.close()
def fileup(x):  #uploading data in csv
    data=[]
    for i in x:
        data.append(i[0:5])
    file=open("d:\\project_train.csv","a",newline="")
    writer=csv.writer(file)
    writer.writerows(data)
    print("FILE STORED ON SYSTEM.")
    file.close()
def passcode():  #ALPHA-NUMERIC CODE CREATOR
    a=str(random.randint(1,9))
    b=chr(random.randint(65,90))
    c=str(random.randint(1,9))
    d=chr(random.randint(65,90))
    e=str(random.randint(0,9))
    f=chr(random.randint(65,90))
    x=a+b+c+d+e+f
    return x;
def crypt(x):  #encrypter
    z=passcode()
    y=passcode()
    x=z+x+y
    return(x)
def decrpt(x): #decrypter
    x=x[6:12]
    return(x)
def reconnect():
    con1= sql.connect(host="db4free.net", user="project_train",password="4627a7eb",database="kv_express2021")
    cur1=con1.cursor()
    print("onlinehost RECONNECTED")
print("ALL FUNCTIONS ARE WORKING PROPERLY :)")
```

```python
#booking block
greet("WELCOME TO THE BOOKING BLOCK")
database=[]
filename()
y=1
count=1
while y!=0:
    print()
    record=[]
    print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX")
    print(count)
    name=input("Enter PASSENGER NAME:- ")
    seatno=int(input("ENTER HIS SEAT NO. :- "))
    type=seat_type(seatno)
    age=int(input("PASSENGER's AGE :- "))
    mail=input("PLEASE ENTER HIS EMAIL ID:- ")
    pwd=passcode()
    pwd_crpt=str(crypt(pwd))
    record.append(seatno)
    record.append(type)
    record.append(name)
    record.append(age)
    record.append(mail)
    now=datetime.now()
    record.append(now.strftime('%Y-%m-%d %H:%M:%S'))
    record.append(pwd_crpt)
    query="insert into train(seat_no,type,name,age,gmail_id,time,pwd) values (%s,%s,%s,%s,%s,%s,%s)"
    values=record
    try:  #sending to localhost
        cur.execute(query,values)
        con.commit()
    except sql.Error as err:
        print("\n!! LOCAL HOST ERROR PLEASE CHECK VALUES & RE-ENTER DATA!!".format(err))
    
    try:  #sending online host
        cur1.execute(query,values)
        con1.commit()
    except sql.Error as err:
        print("\n!! ONLINE HOST ERROR PLEASE CHECK VALUES & RE-ENTER DATA!!".format(err))
    database.append(record)
    print("\nXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX\n")
    y=int(input("END(0) OR CONTINUE(Any key) :- "))
    if(y==0):
        x=input("TYPE 'QUIT' to exit or press any key to continue :- ")
        if(x=='QUIT' or x=="quit" or x=="Quit"):
            y=0
        else:
            y=11
    count+=1
fileup(database)
print("PROCEEDING TOWARDS BOOKING BLOCK.....\n")
sleep(1)
gmails=[]
keys=[]
stype=[]
passenger=[]
seatno=[]
cur.execute("SELECT Name,seat_no,type,PWD,Gmail_Id FROM train where status='NO';")
data_set= cur.fetchall()
for i in data_set:
    passenger.append(i[0])
    seatno.append(str(i[1]))
    stype.append(i[2])
    keys.append(decrpt(i[3]))
    gmails.append(i[4])   
gmail_user = input("ENTER COMPANY's GMAIL-Id:- ")
gmail_password=getpass.getpass("PLEASE ENTER PASSKEY:- ")
greet("MAIL SENDING INITIATED")
t=len(gmails)
for i in range(t): 
    message="RESPECTED "+passenger[i]+".\nYOUR SEAT VERIFICATION KEY FOR SEAT NO. '"+seatno[i]+" - "+stype[i]+"' IS '"+keys[i]+"'.\nUSE THIS CODE ON-BOARD TO SEE YOUR DETAILS & TRAIN SUMMARY.\nPLEASURE TO MEET-YOU :) \nTHANK-YOU\nREGARDS- KV EXPRESS STAFF.\n:)"
    msg = MIMEText(message)
    msg['Subject'] = 'KV EXPRESS'
    msg['From'] = gmail_user
    msg['To'] = gmails[i]

    server = smtplib.SMTP('smtp.gmail.com:587')
    server.starttls()
    server.ehlo()
    server.login(gmail_user, gmail_password)
    server.sendmail(gmail_user, gmails[i], msg.as_string())
    server.quit()
    print(int((i*100)/(t)),"%")
    a=int(seatno[i])
    query="UPDATE train SET status='YES' where seat_no={};".format(a)
    cur.execute(query)
    cur1.execute(query)
con.commit()
print("100%\nALL GMAILS SENT....")
greet("all mail's sent successfully")
```
